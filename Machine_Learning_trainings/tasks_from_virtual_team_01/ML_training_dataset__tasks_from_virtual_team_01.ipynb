```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('ML_training_dataset__tasks_from_virtual_team_01.csv')
```

```python
df
```

```python
df.hist()
```

```python
plt.scatter(x=df.days_to_end_from_planning, y=df.real_day_to_end_after_work_done
, alpha=0.4)
```

```python
from sklearn.preprocessing import StandardScaler
scaler_X = StandardScaler()
scaler_y = StandardScaler()
```

```python
X = scaler_X.fit_transform(df.days_to_end_from_planning.values.reshape(-1, 1))
y = scaler_y.fit_transform(df.real_day_to_end_after_work_done.values.reshape(-1, 1))
```

```python
plt.scatter(x=X, y=y, alpha=0.4)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_lin = LinearRegression() 
reg_lin.fit(X_train, y_train)
```

```python
y_pred = reg_lin.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
X
```

```python
X_poly
```

```python
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.33, random_state=42) 
```

```python
reg_poly = LinearRegression()
reg_poly.fit(X_train_poly, y_train)
```

```python
hlp.plot_regression(X=X_test_poly, y=y_test, reg=reg_poly, poly=poly, title='Test set - PolynomialFeatures')
```

```python
# jakby ot bylo na zbuorze treningowym
hlp.plot_regression(X=X_train_poly, y=y_train, reg=reg_poly, poly=poly, title='Train set - PolynomialFeatures')
```

```python
for d in [1, 2, 3, 4, 8, 18]:
    poly = PolynomialFeatures(degree=d)
    X_poly = poly.fit_transform(X)
    X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.33, random_state=42)
    reg_poly = LinearRegression()
    reg_poly.fit(X_train_poly, y_train)
    # jakby ot bylo na zbuorze treningowym
    hlp.plot_regression(X=X_train_poly, y=y_train, reg=reg_poly, poly=poly, title=f'Train set - PolynomialFeatures - degree {d}')
    
```

```python
d = 4
poly = PolynomialFeatures(degree=d)
X_poly = poly.fit_transform(X)
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.33, random_state=42)
reg_poly = LinearRegression()
reg_poly.fit(X_train_poly, y_train)
# jakby ot bylo na zbuorze treningowym
hlp.plot_regression(X=X_train_poly, y=y_train, reg=reg_poly, poly=poly, title=f'Train set - PolynomialFeatures - degree {d}')
```
